```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
df = sns.load_dataset("tips")
df
```

```python
df.shape
df.info()
```

```python
df.describe()
```

# Análisis univariado
Ayuda a determinar una variable a la vez, nos permite analizar el comportamiento de cada variable.

```python
sns.histplot(data=df, x="total_bill")
plt.plot()
```

```python
sns.boxplot(data=df, x="total_bill")
plt.plot()
```

```python
sns.kdeplot(data=df, x="total_bill")
plt.plot()
```

# Análisi bivariado
Ayuda a ver la ralación entre dos variables

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

```python
sns.barplot(data=df, x="sex", y="tip")
plt.show()
```

```python
sns.boxplot(data=df, x="sex", y="total_bill")
plt.show()
```

```python
sns.pairplot(data=df, hue="sex",)
plt.show()
```
