```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
df = sns.load_dataset("tips")
df
```

```python
df.shape
df.info()
```

```python
df.describe()
```

# Análisis univariado
Ayuda a determinar una variable a la vez, nos permite analizar el comportamiento de cada variable.

```python
sns.histplot(data=df, x="total_bill")
plt.plot()
```

```python
sns.boxplot(data=df, x="total_bill")
plt.plot()
```

```python
sns.kdeplot(data=df, x="total_bill")
plt.plot()
```

# Análisi bivariado
Ayuda a ver la ralación entre dos variables

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

```python
sns.barplot(data=df, x="sex", y="tip")
plt.show()
# Los hombres dan mas propina
```

```python
sns.boxplot(data=df, x="sex", y="total_bill")
plt.show()
```

```python
sns.pairplot(data=df, hue="sex",)
plt.show()
```

# Análisis de los datos con histogramas

```python
# Gráfico de barras de propinas promedio por sexo
sns.barplot(data=df, y="tip", hue ="sex",errorbar=None)
plt.title("Propinas promedio por sexo")
plt.show()

# Los hombres dan mas propinas que los hombres
```

```python
sns.barplot(data=df, hue="day",y="tip",errorbar=None)
plt.title("Propinas promedio por dia")
plt.show()

# Los domingos dan mayor propinas
```

```python
sns.barplot(data=df, hue="time",y="tip",errorbar=None)
plt.title("Propinas promedio por tiempo")
plt.show()

# En la cena dan más propinas
```

```python
# Propinas por fumador
sns.barplot(data=df, hue="smoker",y="tip",errorbar=None)
plt.title("Propinas promedio por Fumador/No fumador")
plt.show()

# el hecho de que tu cliente no afecta a la propina
```

```python
# Graficos de barras de propinas promedio (Dia, sexo)
sns.barplot(data=df, x="day",y="tip", hue="sex",errorbar=None)
plt.title("Propinas promedio por dia y sexo")
plt.show()
```

```python
# Graficos de barras de propinas tiempo
sns.barplot(data=df, x="time",y="tip", hue="sex",errorbar=None)
plt.title("Propinas promedio por tiem´po y sexo")
plt.show()
```

```python
# Graficos de barras de propinas fumador/no fumador
sns.barplot(data=df, x="smoker",y="tip", hue="sex",errorbar=None)
plt.title("Propinas promedio por Fumador/No fumador")
plt.show()
```

```python
# Gráfico de dispersión entre sixe y tip
sns.scatterplot(data= df, x="size", y="tip")
plt.title("Relación entre tamaño de fiesta y propinas")
plt.show()
```

```python
# Gráfico de barras de propinas promedio por tamaño de fiesta
sns.barplot(data=df,y="tip",hue="size",errorbar=None)
plt.title("Propinas promedio por dia y sexo")
plt.show()
```

```python
# Calcul el % de propina y lo agregamos al DataFrame
df["tip_percentage"] = (df["tip"]/df["total_bill"])*100
df
```

```python
# Gráfico de barras de porcentaje de propinas por sexo
sns.barplot(data=df, x="sex", y="tip_percentage", errorbar= None)
plt.title("Porcentaje de propina promedio por sexo")
plt.show()
```

```python
# Gráfico de porcentaje de propina promedio por dia
sns.barplot(data=df, x="day", y="tip_percentage", errorbar= None)
plt.title("Porcentaje de propina promedio por dia")
plt.show()
```

```python
# Gráfico de porcentaje de propina promedio por tiempo
sns.barplot(data=df, x="time", y="tip_percentage", errorbar= None)
plt.title("Porcentaje de propina promedio por tiempo")
plt.show()
```

```python
# Gráfico de porcentaje de propina promedio por si es fumador/ no fumador
sns.barplot(data=df, x="smoker", y="tip_percentage", errorbar= None)
plt.title("Porcentaje de propina promedio por Fumador/No fumador")
plt.show()
```
