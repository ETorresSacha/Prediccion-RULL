```python
import numpy as np
```

```python
# Crear arreglo de 1 dimensión con 5 elementos
arr1 = np.array([1,2,3,4,5,6])
print('Arreglo 1D:', arr1)
print('Forma (shape):',arr1.shape)
print('Tipo de datos (dtype):',arr1.dtype)
```

```python
# Cambiar la forma del arreglo a una matriz 2x3
arr2 =arr1.reshape(2,3) # fila, columna
print('Arreglo 2D:\n', arr2)
print('Forma (shape):',arr2.shape)
print('Tipo de datos (dtype):',arr2.dtype)
```

```python
# Crear una matriz con forma (4,4) con números aleatorios entre 0 y 1.
matrix = np.random.rand(4,4)
print('Arreglo 2D:\n', matrix)
print('Forma (shape):',matrix.shape)
print('Tipo de datos (dtype):',matrix.dtype)
```

```python
import random

# Generar un número aleatorio entre 0 y 1
num_aleatorio = random.random()
print(f"random.random:Número aleatorio entre 0 y 1: {num_aleatorio}")

# Generar un entero aleatorio entre 1 y 10
numero_entero = random.randint(1, 10)
print(f"random.randint:Entero aleatorio entre 1 y 10: {numero_entero}")

# Generar un número flotante aleatorio entre 5.0 y 10.0
numero_flotante = random.uniform(5.0, 10.0)
print(f"random.uniform:Número flotante aleatorio entre 5.0 y 10.0: {numero_flotante}")

# Seleccionar un elemento aleatorio de una lista
colores = ["rojo", "azul", "verde"]
color_elegido = random.choice(colores)
print(f"random.choice:Color elegido aleatoriamente: {color_elegido}")

# Mezclar una lista
numeros = [1, 2, 3, 4, 5]
random.shuffle(numeros)
print(f"random.shuffle:Lista mezclada: {numeros}")

# Seleccionar 2 elementos aleatorios sin reemplazo de una lista
elementos = ['a', 'b', 'c', 'd', 'e']
muestra = random.sample(elementos, 2)
print(f"random.sample:Muestra aleatoria de elementos: {muestra}")

# Inicializar el generador de números aleatorios para reproducibilidad
random.seed(42)
print(f"random.seed:Número aleatorio con semilla 42: {random.random()}")
```
